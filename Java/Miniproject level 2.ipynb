### Main()

```java
public class Main {
    public static void main(String[] args) {
        //Creates new scanner object to recieve user inputs.
        Scanner scanner = new Scanner(System.in);

        //Welcomes the user to the game. 
        System.out.println("Welcome to the Pet Game!");

        //Calls the create new pet method and the play game method.
        String[] pet = createNewPet(scanner);
        playGame(pet, scanner);

        //Closes the scanner
        scanner.close();
    }
//END MAIN
```

This method is called by the main method and it servers the purpose to create a new array containing pet information.
```java
    //New method called createNewpet that accepts the scanner object as a variable. 
    public static String[] createNewPet(Scanner scanner) {
        // Creates new array to holds these values:{name, type, hunger, happiness, health, alive}
        String[] pet = new String[6];

        //Asks the user the name and species of the pet. 
        System.out.print("Please enter your pet's name: ");
        pet[0] = scanner.nextLine();

        System.out.print("Please enter your pet's species: ");
        pet[1] = scanner.nextLine();

        //Sets all of the pets variable values. 
        pet[2] = "50";  // hunger
        pet[3] = "50";  // happiness
        pet[4] = "100"; // health
        pet[5] = "true";  // alive

        return pet;
    }
//END CreateNewPet
```

### PlayGame()

#### Functionality

This is the main part of my program that enables the game to be played. The function defines a set of options for the user so that they can manipulate and look after their own pet. Depending on the option chosen the function will pick out the function it correlates to and will call that method. This function loops until the pet has passed away. 

#### Implementation 

Uses array statements to get the pets details, uses while loops within while loops and if else statements. 


```java
  //Creates a new method that accepts the array pet and the scanner object. 
  public static void playGame(String[] pet, Scanner scanner) {
      boolean quit = false;

      //Creates a new while loop that runs aslong as the pet is not dead or quit is not selected. 
      while (!quit && Boolean.parseBoolean(pet[5])) {

          //Outputs a basic option menu for the user to select a function. 
          System.out.println("1. Feed " + pet[0]);
          System.out.println("2. Play with " + pet[0]);
          System.out.println("3. Take " + pet[0] + " to the vet");
          System.out.println("4. Quit");
          System.out.print("Enter your choice: ");

          //Choices stores the user selection and valid input checks if the input is valid. 
          int choice = -1;
          boolean validInput = false;

          // While Loop to validate player's input which continues while the input is not valid. 
          while (!validInput) {
              if (scanner.hasNextInt()) {
                  //Reads the player input. If its true it sets valid input to true and the loop ends.
                  choice = scanner.nextInt();
                  scanner.nextLine();
                  validInput = true;
              } else {
                  //If there is an invalid input it will prompt the user to re-enter the number. 
                  System.out.println("Invalid input. Please enter a number.");
                  scanner.nextLine();
                  System.out.print("Enter your choice: "); 
              }
          }

          //Uses decision statements to call the functions that manipulate the pet statistics. 
          if (choice == 1) {
              feedPet(pet);
          } else if (choice == 2) {
              playWithPet(pet);
          } else if (choice == 3) {
              takeToVet(pet);
          } else if (choice == 4) {
              quit = true;
              System.out.println("Thanks for Playing!");
          } else {
              System.out.println("Invalid choice.");
          }

          //Checks each round if the pet is still alive. 
          petStatus(pet);
      }
  }
//END PlayGame
```
